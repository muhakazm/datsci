## What kind of data does pandas handle?

- `pandas.Series(data=None, index=None, dtype=None, name=None, copy=None, fastpath=False)`
- `pandas.DataFrame(data=None, index=None, columns=None, dtype=None, copy=None)`

The community agreed alias for pandas is `pd`, so loading `pandas` as `pd` is assumed standard practice for all of the pandas documentation.

```python
import pandas as pd
```

To manually store data in a table, create a `DataFrame`. When using a Python dictionary of lists, the dictionary keys will be used as column headers and the values in each list as columns of the `DataFrame`.

I want to store passenger data of the Titanic. For a number of passengers, I know the name (characters), age (integers) and sex (male/female) data.

```python
df = pd.DataFrame(
    {
        "Name": [
            "Braund, Mr. Owen Harris",
            "Allen, Mr. William Henry",
            "Bonnell, Miss. Elizabeth",
        ],
        "Age": [22, 35, 58],
        "Sex": ["male", "male", "female"],
    }
)
```

```python
df
```

A `DataFrame` is a 2-dimensional data structure that can store data of different types (including characters, integers, floating point values, categorical data and more) in columns. It is similar to a spreadsheet, a SQL table or the `data.frame` in R.

- The table has 3 columns, each of them with a column label. The column labels are respectively `Name`, `Age` and `Sex`.

- The column `Name` consists of textual data with each value a string, the column `Age` are numbers and the column `Sex` is textual data.

**Each column in a `DataFrame` is a `Series`**

When selecting a single column of a pandas `DataFrame`, the result is a pandas `Series`. To select the column, use the column label in between square brackets [].

```python
df["Age"]
```

```python
type(df["Age"])
```

You can create a Series from scratch as well:

```python
ages = pd.Series([22, 35, 58], name="Age")
ages
```

A pandas `Series` has no column labels, as it is just a single column of a `DataFrame`. A `Series` does have row labels.

**Do something with a `DataFrame` or `Series`**

*I want to know the maximum Age of the passengers*

We can do this on the `DataFrame` by selecting the `Age` column and applying `max()`.

Catatan: method [`pandas.Series.max`](https://pandas.pydata.org/docs/reference/api/pandas.Series.max.html) dapat pula dipakai untuk `pandas.DataFrame`; ada beberapa argument lain yang dapat digunakan: `axis`, `skipna`, dan `numeric_only`.

```python
df["Age"].max()
```

```python
ages.max()
```

As illustrated by the `max()` method, you can do things with a `DataFrame` or `Series`. `pandas` provides a lot of functionalities, each of them a method you can apply to a `DataFrame` or `Series`. As methods are functions, do not forget to use parentheses `()`.

*I’m interested in some basic statistics of the numerical data of my data table*

```python
df.describe()
```

The `describe()` method provides a quick overview of the numerical data in a `DataFrame`. As the `Name` and `Sex` columns are textual data, these are by default not taken into account by the `describe()` method.

Many `pandas` operations return a `DataFrame` or a `Series`. The `describe()` method is an example of a pandas operation returning a `pandas` `Series` or a `pandas` `DataFrame`.

Catatan `describe()`:
- Arguments dan default arguments: `describe(percentiles=None, include=None, exclude=None)`
- Data `NA` tidak diperhitungkan
- Argument `percentiles` dipakai untuk mendapatkan persentil yang diinginkan; by default, the median is always included
- Argument `include` dan `exclude` seputar pemilihan kolom dari `DataFrame` mana saja yang akan diproses sesuai dengan tipe datanya

**Remember:**
- Import the package: `import pandas as pd`
- A table of data is stored as a `pandas` `DataFrame`
- Each column in a `DataFrame` is a `Series`
- You can do things by applying a method to a `DataFrame` or `Series`

## How do I select a subset of a table?



### 1. Selecting specific columns from a `DataFrame`

To select a single column, use square brackets [] with the column name of the column of interest.

```python
passengerNames = titanic['Name']
passengerNames
```

Reminder: each column in a `DataFrame` is a `Series`

```python
type(passengerNames)
```

Use the [`shape`](https://pandas.pydata.org/docs/reference/api/pandas.DataFrame.shape.html) attribute for getting the pairing of the number of rows and the number of columns; it returns a tuple representing the dimensionality of the DataFrame.

A pandas Series is 1-dimensional and only the number of rows is returned.

```python
print(titanic.shape)
print(passengerNames.shape)
```

To select multiple columns, put the column names inside a list.

```python
passengerNamesAges = titanic[['Name', 'Age']]
passengerNamesAges
```

### 2. Selecting specific rows from a `DataFrame`

I am interested in the passengers older than 35 years old.

To select rows based on a conditional expression, use a condition inside the selection brackets `[]`.

```python
above_35 = titanic[titanic["Age"]>35]
above_35
```

The condition inside the selection brackets `titanic["Age"] > 35` checks for which rows the `Age` column has a value larger than 35.

```python
titanic["Age"] > 35
```

The output of the conditional expression (`>`, but also `==`, `!=`, `<`, `<=`,… would work) is actually a pandas `Series` of boolean values (either True or False) with the same number of rows as the original `DataFrame`. Only rows for which the value is `True` will be selected.

Reminder: suatu `Series` (atau `list`) dari sejumlah elemen `boolean` dapat digunakan sebagai filter untuk `DataFrame` apabila jumlah elemennya sama dengan jumlah baris dari `DataFrame`.

Banyaknya data (jumlah baris) di `above_35` akan merepresentasikan jumlah penumpang dengan usia lebih dari 35 tahun; untuk itu attribute `shape` dapat digunakan kembali.

```python
above_35.shape
```

I’m interested in the Titanic passengers from cabin class 2 and 3.

Untuk itu, method [`Series.isin()`](https://pandas.pydata.org/docs/reference/api/pandas.Series.isin.html#pandas.Series.isin) dapat digunakan dengan input berupa `list`.

```python
class_23 = titanic[titanic["Pclass"].isin([2, 3])]
class_23
```

The above is equivalent to filtering by rows for which the class is either 2 or 3 and combining the two statements with an `|` (or) operator:

```python
class_23 = titanic[(titanic["Pclass"] == 2) | (titanic["Pclass"] == 3)]
class_23
```

Reminder: when combining multiple conditional statements, each condition must be surrounded by parentheses `()`. Moreover, you can not use `or`/`and` but need to use the or operator `|` and the and operator `&`.

```python
class_23_male = titanic[(titanic["Pclass"] == 2) | (titanic["Pclass"] == 3) & (titanic["Sex"] == 'male')]
class_23_male
```

I want to work with passenger data for which the age is unknown; it can be done with [`isna()`](https://pandas.pydata.org/docs/reference/api/pandas.Series.isna.html#pandas.Series.isna) method.

```python
age_is_na = titanic[titanic["Age"].isna()]
age_is_na
```

I want to work with passenger data for which the age is known; it can be done with [`notna()`](https://pandas.pydata.org/docs/reference/api/pandas.Series.notna.html#pandas.Series.notna) method.

```python
age_no_na = titanic[titanic["Age"].notna()]
age_no_na
```

### 3. Selecting specific rows and columns from a `DataFrame`

I'm interested in the names of the passengers older than 35 years.

```python
adult_names = titanic.loc[titanic["Age"] > 35, "Name"]
adult_names
```

In this case, a subset of both rows and columns is made in one go and just using selection brackets `[]` is not sufficient anymore. The `loc`/`iloc` operators are required in front of the selection brackets `[]`. When using `loc`/`iloc`, the part before the comma is the rows you want, and the part after the comma is the columns you want to select.

#### Berikut contoh berbagai kasus penggunaan property `loc`

Meng-copy 10 baris pertama `DataFrame` `titanic` untuk penyederhanaan:

```python
titanic_10 = titanic.head(10).copy(deep=True) # default-nya memang True
titanic_10
```

Iseng ubah data; karena argument `deep=True`, perubahan pada `DataFrame` hasil `copy` tidak akan mengubah `DataFrame` original.

```python
titanic_10.at[0,'Name'] = 'Braund, Mr. George Owen Harrison'
titanic_10
```

Input (satu buah) integer `n` untuk mendapatkan data ber-index `n`, contohnya: `titanic.loc[0]`.

```python
titanic_10.loc[0]
```

Ingat bahwa index suatu baris berupa `string`.

```python
string_index = [f'row_{i+1}' for i in range(len(titanic_10))] # buat list untuk index baru sejumlah banyaknya baris
titanic_10_strind = titanic_10.copy() # copy DataFrame
titanic_10_strind.index = string_index # pakai list index baru
titanic_10_strind
```

Contoh index string:

```python
titanic_10_strind.loc['row_1']
```

Jika input pada `loc` merupakan list, maka akan menghasilkan `DataFrame`.

```python
titanic_10.loc[[0]]
```

Mengambil sebagian baris berurut tanpa step; seluruh kolom

```python
titanic_10.loc[:4]
```

Mengambil sebagian baris berurut dengan step; seluruh kolom

```python
titanic_10.loc[:8:3]
```

Mengambil seluruh baris berurut; sebagian kolom berurut tanpa step

```python
titanic_10.loc[:,'PassengerId':'Age']
```

Mengambil seluruh baris berurut; sebagian kolom berurut dengan step

```python
titanic_10.loc[:,'PassengerId':'Embarked':2]
```

Mengambil sebagian baris acak; seluruh kolom

```python
titanic_10.loc[[5,1,3,2]]
```

Mengambil seluruh baris; sebagian kolom acak

```python
titanic_10.loc[:,['PassengerId','Name','Sex']]
```

Data penumpang perempuan (`Sex=female`); berusia 30 tahun kebawah (`Age<=30`) atau penumpang kelas 1 (`Pclass=1`)

_Note_: setiap logical unit harus di dalam kurung

```python
titanic_10.loc[((titanic_10["Age"] <= 30) | (titanic_10["Pclass"] == 1)) & (titanic["Sex"] == 'female'), :]
```
